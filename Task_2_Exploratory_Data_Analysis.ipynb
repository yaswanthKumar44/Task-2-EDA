```python
#  Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from google.colab import files
```

```python
# Load dataset
df = pd.read_csv('Titanic.csv')
```

```python
# Generate Summary Statistics
print("ðŸ“Œ Dataset Shape:", df.shape)
print(df.describe())
print(df.info())
```

```python
#  Check for missing values
print("\nðŸ“Œ Missing Values:\n", df.isnull().sum())
```

```python
#  Histograms for Numeric Features
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']
```

```python
plt.figure(figsize=(12,8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()
```

```python
#  Boxplots for Numeric Features
plt.figure(figsize=(12,8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
#  Pairplot for Relationships (Optional â€” works better on smaller samples)
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']])
plt.show()
```

```python
#  Correlation Matrix
plt.figure(figsize=(10,7))
# Select only numeric columns before calculating correlation
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#  Identify Patterns, Trends, or Anomalies
# Survival Rate by Gender
plt.figure(figsize=(6,4))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Survival Rate by Passenger Class
plt.figure(figsize=(6,4))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python

#  Basic Inferences from Visuals
# Fare distribution by survival
fig = px.box(df, x='Survived', y='Fare', color='Survived', title='Fare Distribution by Survival')
fig.show()
```

```python
# Age distribution by survival
fig = px.histogram(df, x='Age', color='Survived', nbins=30, title='Age Distribution by Survival')
fig.show()
```
